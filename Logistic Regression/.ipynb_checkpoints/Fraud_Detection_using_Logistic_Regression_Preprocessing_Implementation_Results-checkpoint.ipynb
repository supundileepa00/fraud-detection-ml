# Fraud Detection using Logistic Regression: Preprocessing, Implementation, and Results

## Import necessary libraries

```python
import pandas as p_d
import numpy as n_p
from sklearn.preprocessing import LabelEncoder as l_e 
from sklearn.preprocessing import MinMaxScaler as m_m_s
from sklearn.model_selection import train_test_split as trainTestSplit
from sklearn.linear_model import LogisticRegression as l_g
from sklearn.metrics import accuracy_score as accuracyScore, precision_score as precisionScore, recall_score as recallScore, f1_score as f1Score
from sklearn.preprocessing import StandardScaler as s_s
from sklearn.metrics import confusion_matrix as c_m
from sklearn.metrics import roc_curve, roc_auc_score
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
from sklearn.calibration import calibration_curve
import matplotlib.pyplot as plt
from sklearn.metrics import precision_recall_curve
import seaborn as sns
```

## Load the dataset

```python
# Load the train dataset into a Pandas DataFrame
transactionDataSet = p_d.read_csv('C:/Users/JanithGamage/Desktop/ML-Assignment/fraudTrain.csv/fraudTrain.csv')

# Display the first few rows of the dataset
transactionDataSet.head()
```

## Data preprocessing  

```python
# Drop unnecessary columns
transactionDataSet = transactionDataSet.drop('Unnamed: 0', axis=1)  # Drop the 'Unnamed: 0' column as it seems to be an unnecessary index

# Convert date columns to datetime format
transactionDataSet['trans_date_trans_time'] = p_d.to_datetime(transactionDataSet['trans_date_trans_time'])
transactionDataSet['dob'] = p_d.to_datetime(transactionDataSet['dob'])

# Extract features from date columns
transactionDataSet['trans_year'] = transactionDataSet['trans_date_trans_time'].dt.year
transactionDataSet['trans_month'] = transactionDataSet['trans_date_trans_time'].dt.month
transactionDataSet['trans_day'] = transactionDataSet['trans_date_trans_time'].dt.day
transactionDataSet['trans_hour'] = transactionDataSet['trans_date_trans_time'].dt.hour
transactionDataSet['dob_year'] = transactionDataSet['dob'].dt.year
transactionDataSet['dob_month'] = transactionDataSet['dob'].dt.month
transactionDataSet['dob_day'] = transactionDataSet['dob'].dt.day

# Drop the original date columns
transactionDataSet = transactionDataSet.drop(['trans_date_trans_time', 'dob'], axis=1)

labelEncoder = l_e()
transactionDataSet['merchant'] = labelEncoder.fit_transform(transactionDataSet['merchant'])
transactionDataSet['category'] = labelEncoder.fit_transform(transactionDataSet['category'])
transactionDataSet['gender'] = labelEncoder.fit_transform(transactionDataSet['gender'])
transactionDataSet['first'] = labelEncoder.fit_transform(transactionDataSet['first'])
transactionDataSet['last'] = labelEncoder.fit_transform(transactionDataSet['last'])
transactionDataSet['street'] = labelEncoder.fit_transform(transactionDataSet['street'])
transactionDataSet['city'] = labelEncoder.fit_transform(transactionDataSet['city'])
transactionDataSet['state'] = labelEncoder.fit_transform(transactionDataSet['state'])
transactionDataSet['job'] = labelEncoder.fit_transform(transactionDataSet['job'])
transactionDataSet['trans_num'] = labelEncoder.fit_transform(transactionDataSet['trans_num'])

scalerMinMax = m_m_s()
transactionDataSet[['amt', 'lat', 'long', 'city_pop', 'merch_lat', 'merch_long']] = scalerMinMax.fit_transform(transactionDataSet[['amt', 'lat', 'long', 'city_pop', 'merch_lat', 'merch_long']])

# Display the first few rows of the preprocessed dataset
transactionDataSet.head()

# Limit the usage of data rows to 100000
transactionDataSet = transactionDataSet.head(100000)

# Save the preprocessed dataset to a CSV file
transactionDataSet.to_csv('C:/Users/JanithGamage/Desktop/ML-Assignment/fraudTrain.csv/preprocessed_fraudTrain_test1.csv', index=False)
```

## LogisticRegression Implementation

```python
# Load the training dataset
preprocessedDataSet = p_d.read_csv('C:/Users/JanithGamage/Desktop/ML-Assignment/fraudTrain.csv/preprocessed_fraudTrain_test1.csv')
 
# Split the data into input features (X) and target variable (y)
XAxisTrain = preprocessedDataSet.drop('is_fraud', axis=1)
yAxisTrain = preprocessedDataSet['is_fraud']

# Split the data into training and testing sets
XAxisTrain, XAxisTest, yAxisTrain, yAxisTest = trainTestSplit(XAxisTrain, yAxisTrain, test_size=0.2, random_state=42)

# Create a s_s object
scaler = s_s()

# Scale the input features (XAxisTrain and XAxisTest)
XAxisTrain_scaled = scaler.fit_transform(XAxisTrain)
XAxisTest_scaled = scaler.transform(XAxisTest)

# Create a Logistic Regression model
model = l_g(max_iter=100000)

# Train the model on the scaled training data
model.fit(XAxisTrain_scaled, yAxisTrain)

# Make predictions on the scaled testing data
yAxisPred = model.predict(XAxisTest_scaled)

# Calculate accuracy, precision, recall, and F1 score
accuracy = accuracyScore(yAxisTest, yAxisPred)
precision = precisionScore(yAxisTest, yAxisPred, zero_division=1)
recall = recallScore(yAxisTest, yAxisPred)
f1_score = f1Score(yAxisTest, yAxisPred)

# Print the evaluation metrics
print('Evaluation Metrics:')
print('-------------------')
print('Accuracy:', accuracy)
print('Precision:', precision)
print('Recall:', recall)
print('F1 Score:', f1_score) 
```

### Graphical l_g Representation 1 - Confusion Matrix:

```python
confusionMatrix = c_m(yAxisTest, yAxisPred)
print(confusionMatrix) 
```

### Graphical l_g representation 2 -  AUC score

```python
# Calculate the probability scores for each instance in the test set
y_proba = model.predict_proba(XAxisTest_scaled)[:, 1]

# Calculate the false positive rate and true positive rate
fpr, tpr, thresholds = roc_curve(yAxisTest, yAxisPred)

# Calculate the AUC score
auc = roc_auc_score(yAxisTest, yAxisPred)

# Plot the ROC curve
plt.plot(fpr, tpr)
plt.plot([0, 1], [0, 1], linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve (AUC = {:.2f})'.format(auc))
plt.show() 
```

### Graphical l_g Representation 3 - Confusion Matrix Diagram

```python
# Calculate the confusion matrix
cm = c_m(yAxisTest, yAxisPred)

# Plot the confusion matrix
sns.heatmap(cm, annot=True, cmap='Blues', fmt='g')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show() 
```

### Graphical l_g Representation 4 - Feature Impotents Plot

```python
importance = model.coef_[0]
features = XAxisTrain.columns
indices = n_p.argsort(n_p.abs(importance))[::-1]

plt.figure(figsize=(10, 6))
plt.bar(range(len(indices)), n_p.abs(importance[indices]))
plt.xticks(range(len(indices)), features[indices], rotation=90)
plt.xlabel('Features')
plt.ylabel('Absolute Coefficient')
plt.title('Feature Importance Plot')
plt.show()
```

### Graphical l_g Representation 5 - Calibration Plot

```python
# Get predicted class probabilities for the scaled testing data
y_prob = model.predict_proba(XAxisTest_scaled)[:, 1]

# Create the calibration plot
fraction_of_positives, mean_predicted_value = calibration_curve(yAxisTest, yAxisPred, n_bins=10)
plt.plot(mean_predicted_value, fraction_of_positives, 's-', label='Logistic Regression')

# Add a reference line for perfect calibration
plt.plot([0, 1], [0, 1], '--', label='Perfect Calibration')

# Set plot labels and legend
plt.xlabel('Mean Predicted Value')
plt.ylabel('Fraction of Positives')
plt.legend()

# Show the plot
plt.show() 
```

### Graphical l_g Representation 6 - Precision Recall Plot

```python
# Get predicted class probabilities for the scaled testing data
y_prob = model.predict_proba(XAxisTest_scaled)[:, 1]

# Calculate precision and recall values at different probability thresholds
precision, recall, thresholds = precision_recall_curve(yAxisTest, yAxisPred)

# Create the precision-recall curve
plt.plot(recall, precision, 's-', label='Logistic Regression')

# Set plot labels and legend
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.legend()

# Show the plot
plt.show()
```
